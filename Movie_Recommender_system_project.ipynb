```python
# checking the format of genres column

movies.iloc[0].genres
```

```python
# we need to change the format to ["Action", "Adventure" , "Fantasy" ,"Science Fiction"]

#since this list is in string format, to convert it
import ast

def convert(object):
  l = []
  for i in ast.literal_eval(object):
    l.append(i['name'])
  return l
```

```python
#testing

convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
#applying the same function on keywords as well to convert the format

movies['keywords'] = movies['keywords'].apply(convert)
```

```python
# for cast column

def convert3(object):
  l = []
  counter = 0
  for i in ast.literal_eval(object):
    if counter != 3:
      l.append(i['name'])
      counter += 1
    else:
      break
  return l
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies['cast']
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
# for crew column : we only want the director of the movie

def fetch_director(object):
  l = []
  for i in ast.literal_eval(object):
    if i['job']=='Director':
      l.append(i['name'])
      break
  return l
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies['crew']
```

```python
# for overview column

movies['overview'][0]

#since this is a string, converting it to a list so that we can concatenate with other lists.
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

Now we will concatenate all the lists to convert into a very big list and then convert that big list again into string so that it will give a paragraph which will be labelled as **Tags**

```python
# removing spaces

movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
# creating tags column and concatenating other columns

movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
# creating a new dataset that has only 3 columns

new_df = movies[['movie_id' , 'title' , 'tags']]
new_df
```

```python
# converting the tags column to string

new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
new_df
```

```python
# converting to lower case

new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
new_df
```

Text Vectorization: Using Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
!pip install nltk
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# converting tags into vectors

cv = CountVectorizer(max_features = 5000 , stop_words = 'english')

cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()

vectors[0]
```

```python
#stemming

ps = PorterStemmer()

def stem(text):
  y = []

  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
#testing

stem('in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. action adventure fantasy sciencefiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d samworthington zoesaldana sigourneyweaver jamescameron')
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
# calculating distances or similarities of movie 1 with others and so on

similarity = cosine_similarity(vectors)
similarity
```

```python
similarity[7]
```

```python
# this gives the most similar movies one after the other.

sorted(list(enumerate(similarity[0])),reverse = True,key = lambda x:x[1])[1:6]
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse=True , key = lambda x:x[1])[1:6]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
